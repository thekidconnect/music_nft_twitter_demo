# **objective**

## we would like to understand which of an artist's followers have ethereum wallets, to ascertain how and to whom the artist might market their nfts.

# **methodologies**

## we pulled a table containing 10,000 of Keyboard Kid's over 15,000 followers, and joined this with another table mapping 15k twitter accounts to ens domain names and ethereum wallet addresses.

```python
# read in Keyboard Kid followers

keyboard_kid_twitter_followers_df = pd.read_csv(data_directory+"keyboard_kid_followers_twitter.csv", index_col=[0]).drop(columns=['Unnamed: 0']).rename(columns={'id': 'twitter_id'})
```

```python
keyboard_kid_twitter_followers_df
```

```python
# join followers to twitter-ens mapping

keyboard_kid_twitter_followers_ens_df = keyboard_kid_twitter_followers_df.merge(twitter_ens_df, on='twitter_id', how='left')
```

```python
# filter for followers with wallet matchede

keyboard_kid_twitter_followers_w_ens_df = keyboard_kid_twitter_followers_ens_df[keyboard_kid_twitter_followers_ens_df['owner_wallet'].notnull()]
```
